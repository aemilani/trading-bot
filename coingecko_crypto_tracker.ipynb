```python
import time
import requests
import numpy as np
import pandas as pd
```

```python
def all_crypto():
    url = "https://api.coingecko.com/api/v3/coins/list"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        
        return data
    except requests.exceptions.HTTPError as e:
        print("Error retrieving data:", e)
        return None
```

```python
def top_crypto(n=100):
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        "vs_currency": "usd",
        "order": "market_cap_desc",
        "per_page": n,
        "page": 1,
        "sparkline": False,
    }
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()
        
        return data
    except requests.exceptions.HTTPError as e:
        print("Error retrieving data:", e)
        return None
```

```python
stats = {}
for crypto in top_crypto_list:
    elem = {}
    crypto_id = crypto['id']
    crypto_symbol = crypto['symbol']
    url = f'https://api.coingecko.com/api/v3/coins/{crypto_id}'
    while True:
        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            break
        except requests.exceptions.HTTPError as e:
            print("Error retrieving data:", e)
            print('Sleeping for 10 seconds ...')
            time.sleep(10)
    elem['twtr_users'] = data['community_data']['twitter_followers']
    elem['tlgm_users'] = data['community_data']['telegram_channel_user_count']
    elem['forks'] = data['developer_data']['forks']
    elem['stars'] = data['developer_data']['stars']
    elem['subs'] = data['developer_data']['subscribers']
    elem['total_issues'] = data['developer_data']['total_issues']
    elem['closed_issues'] = data['developer_data']['closed_issues']
    elem['pull_merged'] = data['developer_data']['pull_requests_merged']
    elem['pull_contributors'] = data['developer_data']['pull_request_contributors']
    elem['code_adds_4w'] = data['developer_data']['code_additions_deletions_4_weeks']['additions']
    elem['code_dels_4w'] = data['developer_data']['code_additions_deletions_4_weeks']['deletions']
    elem['commits_4w'] = data['developer_data']['commit_count_4_weeks']
    commits_4w_series = data['developer_data']['last_4_weeks_commit_activity_series']
    elem['commits_4w_per_day_mean'] = np.mean(commits_4w_series)
    elem['commits_4w_per_day_std'] = np.std(commits_4w_series)
    stats[crypto_symbol] = elem
stats = pd.DataFrame.from_dict(stats, orient='index')
stats['code_dels_4w'] = stats['code_dels_4w'].abs()
```
